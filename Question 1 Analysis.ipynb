```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cleaned_data.csv")
```

```python
plt.plot(df.groupby("year")['binned_director'].mean(), label = "Director")
df.groupby("year")['binned_writer'].mean().plot(label = "Writer")
df.groupby("year")['binned_star'].mean().plot(label = "Leading Star")
plt.xlabel("Year")
plt.title("Proportion of Movies with Experienced _______")
plt.ylabel("Proportion")
plt.legend()
```

**Conducted regression on 1990+ data only**

```python
f_df = df[df.year>=1990]
```

```python
f_df['binROI'] = f_df.ROI > 0
```

```python
f_df.genre.value_counts().plot(kind='bar')
plt.xlabel("Genre")
```

```python
f_df.rating.value_counts().plot(kind='bar')
plt.xlabel("Rating")
```

```python
fig, axs = plt.subplots(1,4, sharey=True)
f_df.binned_director.value_counts().plot(kind='bar', ax = axs[0])
axs[0].set_xlabel("Director")
f_df.binned_writer.value_counts().plot(kind='bar', ax = axs[1])
axs[1].set_xlabel("Writer")
f_df.binned_star.value_counts().plot(kind='bar', ax = axs[2])
axs[2].set_xlabel("Lead Star")
f_df.binned_company.value_counts().plot(kind='bar', ax = axs[3])
axs[3].set_xlabel("Company")
plt.tight_layout()
plt.tight_layout()
```

```python
plt.plot(f_df.groupby("year")['binned_director'].mean(), label = "Director")
f_df.groupby("year")['binned_writer'].mean().plot(label = "Writer")
f_df.groupby("year")['binned_star'].mean().plot(label = "Leading Star")
plt.xlabel("Year")
plt.title("Proportion of Movies with Experienced _______")
plt.ylabel("Proportion")
plt.legend()
```
